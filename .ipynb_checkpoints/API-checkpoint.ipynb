```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
client = {
    'username': 'blankhall98',
    'name': 'Jonatan',
    'lastname': 'Blank',
    'password': '1234',
    'age': 24,
    'gender': 'Male',
    'email': 'blankhall@ciencias.unam.mx'
}
```

```python
class API:
    
    def __init__(self,inputs):
        self.comments = []
        self.client = inputs
        self.verify()
        if self.real_client == True:
            self.greet()
            self.menu()
        self.goodbye()
        
    def verify(self):
        self.real_client = False
        print('Please verify your information'+'\n')
        username = input('username: ')
        password = input('password: ')
        if username == self.client['username'] and password == self.client['password']:
            self.real_client = True
            print('Verified!')
        else:
            print('Username or Password doesnt match!')
            
    def greet(self):
        name = self.client['name']
        lastname = self.client['lastname']
        print('API initialized'+'\n'+f'Welcome {name} {lastname}')
        
    def goodbye(self):
        print('Thanks for using my app. Come back soon :D')
        
    def menu(self):
        #SHOW OPTIONS
        print('----------'+'\n')
        print('Select the number of the action to perform:'+'\n')
        print('(1) Get Information')
        print('(2) Post Information')
        print('(3) Edit posted Information')
        print('(4) Delete Information')
        print('(5) Comment!')
        print('\n'+'----------')
        #CLIENT SELECTS OPTION
        do_action = 'yes'
        while do_action == 'yes':
            
            selected_action = input('Action to perform: ')
            #HANDLE CLIENTS SELECTION
            if selected_action == '1':
                self.GET()
            elif selected_action == '2':
                self.POST()
            elif selected_action == '3':
                self.PUT()
            elif selected_action == '4':
                self.DELETE()
            elif selected_action == '5':
                self.COMMENT()
            else:
                print('The option you selected is not correct. Choose again!')
            
            do_action = input('Do another action? (yes/no): ')
            
    def GET(self):
        if len(self.comments) == 0:
            print('Sorry! No comments yet :(')
        else:
            for i in range(len(self.comments)):
                comment = self.comments[i]
                print(f'{i+1} - {comment}')
                
    def POST(self):
        print('POST')
        
    def PUT(self):
        print('PUT')
        
    def DELETE(self):
        print('DELETE')
        
    def COMMENT(self):
        comment = input('write your comment: ')
        print('comment saved!')
        self.comments.append(comment)
```

```python
app = API(client)
```
